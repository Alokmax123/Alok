```python
#6.Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
def calculate_median(data, column_name):
    sorted_data = data.sort_values(by=[column_name])
    n = len(sorted_data)
    if n % 2 == 0:
        median = (sorted_data.iloc[int(n / 2) - 1][column_name] + sorted_data.iloc[int(n / 2)][column_name]) / 2
    else:
        median = sorted_data.iloc[int(n / 2)][column_name]
    return median

median_of_median_income = calculate_median(data, 'median_income')
print(median_of_median_income)
```
